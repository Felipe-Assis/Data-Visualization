```python
def fillColumns(dataframe, columns, value):
    for column in columns:
        dataframe[column] = value
    return dataframe

def getTable(league, season):
    df_season = pd.read_csv('./Data/'+league+'/season-'+convertSeason(season)+'_csv.csv' ,
                            sep = ',', parse_dates=True) 
    
    clubs = df_season['HomeTeam'].unique()
    table = pd.DataFrame(columns=['Name', 'Nicknames', 'Continent', 'Country', 'Founded_Date',
                                  'Season', 'Manager', 'Matches',
                                  'Place', 'Place_Matchday',
                                  'Points', 'Points_Cumulative', 
                                  'W', 'W_Cumulative', 'D', 'D_Cumulative', 'L', 'L_Cumulative',
                                  'Goals_Scored', 'Goals_Scored_Matchday', 'Goals_Scored_Cumulative',
                                  'Goals_Conceded', 'Goals_Conceded_Matchday', 'Goals_Conceded_Cumulative',
                                  'Goal_Difference', 'Goal_Difference_Matchday', 'Goal_Difference_Cumulative',
                                  'Field', 'Opposition', 'Result',
                                  'YellowCards', 'YellowCards_Matchday', 'YellowCards_Cumulative',
                                  'RedCards', 'RedCards_Matchday', 'RedCards_Cumulative',
                                  'Fouls', 'Fouls_Matchday', 'Fouls_Cumulative',
                                  'Squad_Size', 'Squad_Value', 'Average_Age', 'Average_Weight', 'Average_Height',                    
                                  
                                  'Best_Player_Name', 'Best_Player_DateBorn', 'Best_Player_BirthLoc', 
                                  'Best_Player_Height', 'Best_Player_Weight', 'Best_Player_Apps',
                                  'Best_Player_Mins', 'Best_Player_Goals', 'Best_Player_Assists',
                                  'Best_Player_Yel', 'Best_Player_Red', 'Best_Player_SpG', 'Best_Player_PS%',
                                  'Best_Player_AerialsWon', 'Best_Player_MOTM', 'Best_Player_Rating',
                                  'Best_Player_Render', 'Best_Player_Cutout',
                                  
                                  'Has_Stadium', 'Stadium_Name', 'Stadium_Capacity', 'Stadium_Url', 'Logo_Url',
                                  'Points_Cumulative_Average', 'W_Cumulative_Average', 'D_Cumulative_Average', 
                                  'L_Cumulative_Average', 'Goals_Scored_Cumulative_Average', 'Goals_Conceded_Cumulative_Average',
                                  'Goal_Difference_Cumulative_Average', 'YellowCards_Cumulative_Average',
                                  'RedCards_Cumulative_Average', 'Fouls_Cumulative_Average'])
    
    
    
    table['Name'] = clubs
    
    table = fillColumns(table, ['Points', 'W', 'D', 'L', 'Goals_Scored', 'Goals_Conceded', 'Goal_Difference',
                                'YellowCards', 'RedCards', 'Fouls'], 0)
    
    table = fillColumns(table, ['Goals_Scored_Matchday', 'Goals_Scored_Cumulative',
                                    'Goals_Conceded_Matchday', 'Goals_Conceded_Cumulative',
                                    'Goal_Difference_Matchday', 'Goal_Difference_Cumulative',
                                    'YellowCards_Matchday', 'YellowCards_Cumulative',
                                    'RedCards_Matchday', 'RedCards_Cumulative',
                                    'Fouls_Matchday', 'Fouls_Cumulative',
                                    'Field', 'Opposition', 'Result',
                                    'Points_Cumulative', 'Place_Matchday',
                                    'W_Cumulative', 'L_Cumulative', 'D_Cumulative',
                                    'Points_Cumulative_Average', 'W_Cumulative_Average', 'D_Cumulative_Average', 
                                    'L_Cumulative_Average', 'Goals_Scored_Cumulative_Average', 'Goals_Conceded_Cumulative_Average',
                                    'Goal_Difference_Cumulative_Average', 'YellowCards_Cumulative_Average',
                                    'RedCards_Cumulative_Average', 'Fouls_Cumulative_Average'], '')
    
    for i, match in df_season.iterrows():
        homeTeam = match['HomeTeam']
        awayTeam = match['AwayTeam']
        
        # Get clubs index at the table dataframe
        keyH = table.loc[ table['Name'] == homeTeam ].index
        keyA = table.loc[ table['Name'] == awayTeam ].index
        
        #==================================
        #Compute match result
        #==================================
        
        #If Match ends in HOME Team Victory
        if (match['FTR'] == 'H'):
            table.loc[keyH, 'W'] += 1 # Wins counter (home team)
            table.loc[keyA, 'L'] += 1 # Losses counter (away team)  
            table.loc[keyH, 'Points'] += 3 # Points counter (home team)
            table.loc[keyH, 'Result'] += 'W, '
            table.loc[keyA, 'Result'] += 'L, '
            
        # If match ends in AWAY Team Victory
        if (match['FTR'] == 'A'):
            table.loc[keyH, 'L'] += 1 # Losses counter (home team)
            table.loc[keyA, 'W'] += 1 # Wins counter (away team)
            table.loc[keyH, 'L_Cumulative'] += str(table.loc[keyH, 'L'].values[0]) + ',' # Cumulative Wins
            table.loc[keyA, 'W_Cumulative'] += str(table.loc[keyA, 'W'].values[0]) + ',' # Cumulative Losses    
    
            table.loc[keyA, 'Points'] += 3 # Points counter (away team)
            table.loc[keyH, 'Result'] += 'L, '
            table.loc[keyA, 'Result'] += 'W, '
            
        # If match ends in DRAW    
        if (match['FTR'] == 'D'):
            table.loc[keyH, 'D'] += 1 # Draws counter (home team)
            table.loc[keyA, 'D'] += 1 # Draws counter (away team)
 
            
            table.loc[keyH, 'Points'] += 1 # Points counter (home team)
            table.loc[keyA, 'Points'] += 1 # Points counter (away team)
            table.loc[keyH, 'Result'] += 'D,'
            table.loc[keyA, 'Result'] += 'D,'
         
        # Uptade points progression list
        table.loc[keyH, 'W_Cumulative'] += str(table.loc[keyH, 'W'].values[0]) + ',' # Cumulative Wins
        table.loc[keyA, 'W_Cumulative'] += str(table.loc[keyA, 'W'].values[0]) + ',' # Cumulative Losses    
        table.loc[keyH, 'L_Cumulative'] += str(table.loc[keyH, 'L'].values[0]) + ',' # Cumulative Wins
        table.loc[keyA, 'L_Cumulative'] += str(table.loc[keyA, 'L'].values[0]) + ',' # Cumulative Losses  
        table.loc[keyH, 'D_Cumulative'] += str(table.loc[keyH, 'D'].values[0]) + ',' # Cumulative Wins
        table.loc[keyA, 'D_Cumulative'] += str(table.loc[keyA, 'D'].values[0]) + ',' # Cumulative Losses   
        
        table.loc[keyA, 'Points_Cumulative'] += str(table.loc[keyA, 'Points'].values[0]) + ','
        table.loc[keyH, 'Points_Cumulative'] += str(table.loc[keyH, 'Points'].values[0]) + ','
            
            
            
        #Computer Goals Scored
        #Home Team
        table.loc[keyH, 'Goals_Scored'] += match['FTHG'] #Total Home team goals scored
        table.loc[keyH, 'Goals_Scored_Matchday'] += str(match['FTHG'])+', '
        table.loc[keyH, 'Goals_Scored_Cumulative'] += str(table.loc[keyH, 'Goals_Scored'].values[0]) + ','
        #Away Team
        table.loc[keyA, 'Goals_Scored'] += match['FTAG'] #Total Away team goals scored-=-
        
        table.loc[keyA, 'Goals_Scored_Matchday'] += str( match['FTAG'] )+', '
        table.loc[keyA, 'Goals_Scored_Cumulative'] += str(table.loc[keyA, 'Goals_Scored'].values[0]) + ','
        
        
        #Computer Goals Conceded
        #Home Team
        table.loc[keyH, 'Goals_Conceded'] += match['FTAG'] #Total Home team goals conceded
        table.loc[keyH, 'Goals_Conceded_Matchday'] += str(match['FTAG'])+', '
        table.loc[keyH, 'Goals_Conceded_Cumulative'] += str(table.loc[keyH, 'Goals_Conceded'].values[0]) + ','
        #Away Team
        table.loc[keyA, 'Goals_Conceded'] += match['FTHG'] #Total Away team goals conceded
        table.loc[keyA, 'Goals_Conceded_Matchday'] += str( match['FTHG'] )+', '
        table.loc[keyA, 'Goals_Conceded_Cumulative'] += str(table.loc[keyA, 'Goals_Conceded'].values[0]) + ','
        
    
    
        #Compute Goal Difference
        #Home Team
        table.loc[keyH, 'Goal_Difference'] +=  match['FTHG'] - match['FTAG'] 
        table.loc[keyH, 'Goal_Difference_Matchday'] += str(match['FTHG'] - match['FTAG']) + ', '
        table.loc[keyH, 'Goal_Difference_Cumulative'] += str( table.loc[keyH, 'Goal_Difference'].values[0]) + ','
        #Away Team
        table.loc[keyA, 'Goal_Difference'] +=  match['FTAG'] - match['FTHG'] 
        table.loc[keyA, 'Goal_Difference_Matchday'] += str(match['FTAG'] - match['FTHG']) + ', '
        table.loc[keyA, 'Goal_Difference_Cumulative'] += str( table.loc[keyA, 'Goal_Difference'].values[0]) + ','
        
        
        #Compute Yellow Cards
        #Home
        table.loc[keyH, 'YellowCards'] += match['HY']
        table.loc[keyH, 'YellowCards_Matchday'] += str(match['HY']) + ', '
        table.loc[keyH, 'YellowCards_Cumulative'] += str(table.loc[keyH, 'YellowCards'].values[0]) + ','
        #Away
        table.loc[keyA, 'YellowCards'] += match['AY']
        table.loc[keyA, 'YellowCards_Matchday'] += str(match['AY']) + ', '
        table.loc[keyA, 'YellowCards_Cumulative'] += str(table.loc[keyA, 'YellowCards'].values[0]) + ','

        
        #Compute Red Cards
        #Home
        table.loc[keyH, 'RedCards'] += match['HR']
        table.loc[keyH, 'RedCards_Matchday'] += str(match['HR']) + ', '
        table.loc[keyH, 'RedCards_Cumulative'] += str(table.loc[keyH, 'RedCards'].values[0]) + ','
        #Away
        table.loc[keyA, 'RedCards'] += match['AR']
        table.loc[keyA, 'RedCards_Matchday'] += str(match['AR']) + ', '
        table.loc[keyA, 'RedCards_Cumulative'] += str(table.loc[keyA, 'RedCards'].values[0]) + ','
        
        
        #Compute Fols
        #Home
        table.loc[keyH, 'Fouls'] += match['HF']
        table.loc[keyH, 'Fouls_Matchday'] += str(match['HF']) + ', '
        table.loc[keyH, 'Fouls_Cumulative'] += str(table.loc[keyH, 'Fouls'].values[0]) + ','
        #Away
        table.loc[keyA, 'Fouls'] += match['AY']
        table.loc[keyA, 'Fouls_Matchday'] += str(match['AF']) + ', '
        table.loc[keyA, 'Fouls_Cumulative'] += str(table.loc[keyA, 'Fouls'].values[0]) + ','
        
        
        
        #Opposition List
        table.loc[keyH, 'Opposition'] += awayTeam + ','
        table.loc[keyA, 'Opposition'] += homeTeam + ','
        
        #Field
        table.loc[keyH, 'Field'] += 'H,'
        table.loc[keyA, 'Field'] += 'A,'
            
    
    table = setLeagueSeasonInfo(table, league, season)
    table = setPlaces(table.copy())
    table = setPlacesMatchday(table)
    table = setAverages(table)
    return table
```

```python
def setAverages(dataframe):
    champion = dataframe.loc[0] #Drops 1st placed team
    others = dataframe.drop(0)
    parameters = [{'average': 'Points_Cumulative_Average', 'single': 'Points_Cumulative'},
                  {'average': 'W_Cumulative_Average', 'single': 'W_Cumulative'},
                  {'average': 'D_Cumulative_Average', 'single': 'D_Cumulative'},
                  {'average': 'L_Cumulative_Average', 'single': 'L_Cumulative'},
                  {'average': 'Goals_Scored_Cumulative_Average', 'single': 'Goals_Scored_Cumulative'},
                  {'average': 'Goals_Conceded_Cumulative_Average', 'single': 'Goals_Conceded_Cumulative'},
                  {'average': 'Goal_Difference_Cumulative_Average', 'single': 'Goal_Difference_Cumulative'},
                  {'average': 'YellowCards_Cumulative_Average', 'single': 'YellowCards_Cumulative'},
                  {'average': 'RedCards_Cumulative_Average', 'single': 'RedCards_Cumulative'},
                  {'average': 'Fouls_Cumulative_Average', 'single': 'Fouls_Cumulative'}]
    
    
    for parameter in parameters:
            aux = [0] * champion['Matches']
            for j, team in others.iterrows():
                team_values = list(map(float, team[parameter['single']][:-1].split(',')) )
                aux = [x + y for x, y in zip(aux, team_values)]
                
            aux = ','.join(map(str, [np.round(x / others.shape[0],1) for x in aux] )) 
            dataframe.loc[0, parameter['average']] = aux

    return dataframe
```

```python
def setPlacesMatchday(dataframe):
    numMatches = dataframe.iloc[0].Matches
    numTeams = dataframe.shape[0]
    
    for i in range (numMatches):
        auxTable = pd.DataFrame(columns=['Name','Place','Points', 'Goal_Difference', 'Goals_Scored'], index=range(numTeams))
        for j, team in dataframe.iterrows():
            auxTable.loc[j, 'Name'] = dataframe.loc[j, 'Name']
            auxTable.loc[j, 'Points'] = int(dataframe.loc[j, 'Points_Cumulative'].split(',')[:-1][i])
            auxTable.loc[j, 'Goal_Difference'] = int(dataframe.loc[j, 'Goal_Difference_Cumulative'].split(',')[:-1][i])
            auxTable.loc[j, 'Goals_Scored'] = int(dataframe.loc[j, 'Goals_Scored_Cumulative'].split(',')[:-1][i])
            
        auxTable = auxTable.sort_values(by=['Points', 'Goal_Difference', 'Goals_Scored'], ascending=False)
        auxTable = auxTable.reset_index(drop=True)
        for j, team in auxTable.iterrows():
            auxTable.loc[j, 'Place'] = j+1
            
        for j, team in dataframe.iterrows():
            dataframe.loc[j, 'Place_Matchday'] += str(auxTable.loc[auxTable['Name']==team['Name']]['Place'].values[0]) + ','   
            
    return dataframe
```

```python
def setPlaces(dataframe):
    dataframe.sort_values(by=['Points', 'Goal_Difference', 'Goals_Scored'], ascending=False, inplace=True)
    dataframe.reset_index(inplace = True, drop=True) 
    for i, team in dataframe.iterrows():
        dataframe.loc[i, 'Place'] = i+1
        
    return dataframe
```

```python
def setLeagueSeasonInfo(dataframe, league, season):
    aux = getLeagueInfo(league)
    for i, team in dataframe.iterrows():
        clubName = dataframe.loc[i, 'Name'];
        clubInfo = getTeamInfo(clubName)
        
        if (clubInfo != False):
            dataframe.loc[i, 'Logo_Url'] = clubInfo['strTeamBadge']
            dataframe.loc[i, 'Founded_Date'] = clubInfo['intFormedYear']
            dataframe.loc[i, 'Has_Stadium'] = True,
            dataframe.loc[i, 'Stadium_Capacity'] = clubInfo['intStadiumCapacity']
            dataframe.loc[i, 'Stadium_Name'] = clubInfo['strStadium']
            dataframe.loc[i, 'Stadium_Url'] = clubInfo['strStadiumThumb']
                
        dataframe.loc[i, 'Continent'] = aux['Continent']
        dataframe.loc[i, 'Country'] = aux['Country']
        dataframe.loc[i, 'League'] = aux['League']
        dataframe.loc[i, 'Matches'] = aux['Matches']
        dataframe.loc[i, 'Season'] = season
    return dataframe
```

```python
def getLeagueInfo(league):
    if (league=='Bundesliga'):
        return {'Continent': 'Europe', 'Country': 'Germany', 'League': 'Bundesliga', 'Matches': 34}
    if (league=='Premier League'): 
        return {'Continent': 'Europe', 'Country': 'England', 'League': 'Premier League', 'Matches': 38}
    if (league=='La Liga'): 
        return {'Continent': 'Europe', 'Country': 'Spain', 'League': 'La Liga', 'Matches': 38}
    if (league=='Serie A'): 
        return {'Continent': 'Europe', 'Country': 'Italy', 'League': 'Seria A', 'Matches': 38}
    if (league=='Ligue 1'): 
        return {'Continent': 'Europe', 'Country': 'France', 'League': 'Ligue 1', 'Matches': 38}
```
